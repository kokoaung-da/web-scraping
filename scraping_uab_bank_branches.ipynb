# Scraping uab bank Branches

```python
import io
import pandas as pd
import requests 

url = "https://www.uab.com.mm/location/branches/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

try:
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        html_data = io.StringIO(response.text)

        tables = pd.read_html(html_data)

        print(f"There are {len(tables)} tables in this website")

        if len(tables) > 0:
            print("\n--- The First 5 Rows of The First Table ---")
            print(tables[0].head())
    else:
        print(
            f"Fail Connection with Website. Status Code : {response.status_code}"
        )

except Exception as e:
    print(f"Can't extract table. Error : {e}")
```
